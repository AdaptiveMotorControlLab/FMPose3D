# QuickVim

## Functions

### how to search specific text?

To search in vim, follow these steps:

	1.	Press / in normal mode.
	2.	Type the search term.
	3.	Press Enter.

vim will then highlight the first occurrence of the term. You can navigate through the matches with:

	•	n to go to the next match.
	•	N to go to the previous match.

## zsh

```bash
# download oh my zsh
sh -c "$(curl -fsSL https://raw.githubusercontent.com/ohmyzsh/ohmyzsh/master/tools/install.sh)" "" --unattended
# download two 
git clone https://github.com/zsh-users/zsh-syntax-highlighting.git ~/.oh-my-zsh/custom/plugins/zsh-syntax-highlighting

git clone https://github.com/zsh-users/zsh-autosuggestions ~/.oh-my-zsh/custom/plugins/zsh-autosuggestions
```


```bash
# Update ~/.zshrc to enable plugins
sed -i 's/plugins=(git)/plugins=(git zsh-syntax-highlighting zsh-autosuggestions)/' ~/.zshrc

# Optional: Install Powerlevel10k theme
git clone --depth=1 https://github.com/romkatv/powerlevel10k.git ~/.oh-my-zsh/custom/themes/powerlevel10k
sed -i 's/ZSH_THEME="robbyrussell"/ZSH_THEME="powerlevel10k\/powerlevel10k"/' ~/.zshrc

# Reload zsh configuration
source ~/.zshrc
```
