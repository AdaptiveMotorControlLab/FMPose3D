## COCO format

```bash
bbox:[xmin, ymin, width, height]
```

Keypoints:
- Each keypoint is represented by three values:
    - X coordinate: Horizontal position of the keypoint.
    - Y coordinate: Vertical position of the keypoint.
    - Visibility: Indicates whether the keypoint is visible.
        - 0: The keypoint is not labeled (not in the image).
        - 1: The keypoint is labeled but not visible (e.g., occluded).
        - 2: The keypoint is labeled and visible.
        

## SuperAnimal format


Keypoints:
- Each keypoint is represented by three values:
    - X coordinate: Horizontal position of the keypoint.
    - Y coordinate: Vertical position of the keypoint.
    - Visibility: Indicates whether the keypoint is visible.
        - -1: the keypoint is not defined for the dataset from which the image comes from 
        - 0: the keypoint is not visible in the image
        - 2: the keypoint is visible in the image
