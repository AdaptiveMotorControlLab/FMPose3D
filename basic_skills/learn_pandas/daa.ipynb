## use ndarrays to create a new dataframe


```python
import numpy as np
import pandas as pd

ndarray_data = np.array([
    ['Google', 10],
    ['Runoob', 12],
    ['Wiki', 13]
])

df = pd.DataFrame(ndarray_data, columns=['Site', 'Age'])
print(df)

```

## use dict to create a new dataframe


```python
import pandas as pd 

data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'city': ['NY', 'SF', 'LA']
}

df = pd.DataFrame(data)
print(df)

print("______")

data = [{'a': 1, 'b': 2}, {'a': 5, 'b': 10, 'c': 20}]

df = pd.DataFrame(data)

print (df)
```

## loc

```python
import pandas as pd

# Create a sample DataFrame
data = {
    'name': ['Alice', 'Bob', 'Charlie'],
    'age': [25, 30, 35],
    'city': ['NY', 'SF', 'LA']
}
df = pd.DataFrame(data)

# This is what the DataFrame looks like:
#      name  age city
# 0   Alice   25   NY
# 1     Bob   30   SF
# 2  Charlie  35   LA

# Using df.loc[0] will return the first row:
print(df.loc[0])
# Output:
# name    Alice
# age        25
# city       NY
```

```python
# get specific row and column
df.loc[0, 'name']
```

```python
df.loc[0:1]
```

## Retrieve Multiple Rows

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data)

# return the first and second row
print(df.loc[[0, 1]])
```

```python
# set the sepcific index
df = pd.DataFrame(data, index=['day1', 'day2', 'day3'])

print(df)

print("______")

print(df.loc['day2'])
```
