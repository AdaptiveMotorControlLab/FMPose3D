```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Patch

# Configure matplotlib for CVPR paper style (LaTeX rendering)
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'Times', 'DejaVu Serif'],
    'font.size': 10,
    'axes.labelsize': 10,
    'axes.titlesize': 11,
    'xtick.labelsize': 9,
    'ytick.labelsize': 9,
    'legend.fontsize': 9,
    'figure.titlesize': 11,
    'text.usetex': False,  # Set to True if LaTeX is installed
    'mathtext.fontset': 'stix',  # Use STIX fonts for math (closer to LaTeX)
    'mathtext.default': 'regular',
})

# Extract data for different steps
steps = list(range(1, 21))
p1_values = [results_diff_s[i]['p1'] for i in steps]
p2_values = [results_diff_s[i]['p2'] for i in steps]

# Find minimum points
min_p1_idx = p1_values.index(min(p1_values))
min_p2_idx = p2_values.index(min(p2_values))
min_p1_step = steps[min_p1_idx]
min_p2_step = steps[min_p2_idx]

# Create figure with dual y-axes for better readability
fig, ax1 = plt.subplots(1, 1, figsize=(10, 6))

# Highlight optimal region (steps 3-5) - where both metrics are near their minimum
ax1.axvspan(3, 5, alpha=0.15, color='green', zorder=0, label='Optimal Region')

# Plot MPJPE on left y-axis
line1 = ax1.plot(steps, p1_values, marker='o', linewidth=2, markersize=7, 
                 label='MPJPE', color='#1f77b4', zorder=3)
# Mark minimum point with star
ax1.plot(min_p1_step, min(p1_values), marker='*', markersize=11, color='#1f77b4', zorder=4)

ax1.set_xlabel('Step (S)', fontsize=10, fontweight='bold')
ax1.set_ylabel('MPJPE (mm)', fontsize=10, fontweight='bold', color='#1f77b4')
ax1.tick_params(axis='y', labelcolor='#1f77b4', labelsize=9)
ax1.tick_params(axis='x', labelsize=9)
ax1.set_xticks(steps)
ax1.grid(True, alpha=0.3, linestyle='--', zorder=1)

# Create second y-axis for P-MPJPE
ax2 = ax1.twinx()
line2 = ax2.plot(steps, p2_values, marker='s', linewidth=2, markersize=7, 
                 label='P-MPJPE', color='#ff7f0e', zorder=3)
# Mark minimum point with star
ax2.plot(min_p2_step, min(p2_values), marker='*', markersize=11, color='#ff7f0e', zorder=4)

ax2.set_ylabel('P-MPJPE (mm)', fontsize=10, fontweight='bold', color='#ff7f0e')
ax2.tick_params(axis='y', labelcolor='#ff7f0e', labelsize=9)

# Title
# ax1.set_title('MPJPE and P-MPJPE Performance vs Step', fontsize=11, fontweight='bold')

# Combined legend
lines = line1 + line2
labels = [l.get_label() for l in lines]
# Add optimal region to legend
lines.append(Patch(facecolor='green', alpha=0.15))
labels.append('Optimal Region')
ax1.legend(lines, labels, loc='upper right', fontsize=9, framealpha=0.9)

plt.tight_layout()
plt.savefig('images/p1p2_combined_vs_step.jpg', dpi=300, bbox_inches='tight')
plt.show()

# Print statistics
print(f"MPJPE Performance (S1 -> S20):")
print(f"  MPJPE: {p1_values[0]:.4f} -> {p1_values[-1]:.4f} ( = {p1_values[-1] - p1_values[0]:.4f})")
print(f"  Best MPJPE: {min(p1_values):.4f} at Step {min_p1_step}")
print(f"\nP-MPJPE Performance (S1 -> S20):")
print(f"  P-MPJPE: {p2_values[0]:.4f} -> {p2_values[-1]:.4f} ( = {p2_values[-1] - p2_values[0]:.4f})")
print(f"  Best P-MPJPE: {min(p2_values):.4f} at Step {min_p2_step}")
```
